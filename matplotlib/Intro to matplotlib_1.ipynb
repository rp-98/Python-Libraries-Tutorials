**Matplotlib is a multiplatform data visualization library built on NumPy arrays, and designed to work with the broader SciPy 
stack. It was conceived by John Hunter in 2002, originally as a patch to IPython for enabling interactive MATLAB-style 
plotting via gnuplot from the IPython command line**

```python
#importin library
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x= np.linspace(0, 4*(np.pi), 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.grid()
plt.show()
```

Plotting interactively within an IPython notebook can be done with the %matplotlib command, and works in 
a similar way to the IPython shell.

```python
#putting X, Y limpt
p=np.pi
plt.plot(x, np.sin(x))
plt.axis([-p,p,-1,1])    #plt.axis([xmin,xmax,ymin,xmax])
```

```python
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.axis('tight');
plt.subplot(2,1,2)
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
#Labelling the figure
plt.style.use('seaborn-ticks')
plt.plot(x, np.sin(x),'r*')
plt.title("A Sine raph in python")
plt.xlabel("X")
plt.ylabel("sin(x)");
plt.grid()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('tight')
plt.legend();
plt.grid()
```

```python
pi=np.pi
x=np.linspace(-2*pi,2**pi,100)
y1=np.cos(x)
y2=np.sin(x)
plt.plot(x, y1, '-vk')
plt.plot(x,y2,'sr')               #markers --v,s,^,p
```

```python
x1=np.linspace(-2*pi,2**pi,20)
y=np.sin(x1)
plt.plot(x1,y,'-p',markersize=20,markerfacecolor='red',label='sin(X)')
plt.legend()
```

```python
x1=np.linspace(-2*pi,2**pi,50)
y=np.sin(x1)
plt.scatter(x1,y,marker='^',color='red')
```
