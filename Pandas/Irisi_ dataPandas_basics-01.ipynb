# Dealing with categorical features

```python
categorical_features = df2.dtypes==object
categorical_features
```

```python
categorical_cols = df2.columns[categorical_features].tolist()
categorical_cols
```

```python
set(df2['RIVER'])
```

# Label Encoding

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
df2[categorical_cols] = df2[categorical_cols].apply(lambda col: le.fit_transform(col))
```

```python
df2.head()
```

```python
df2.info()
```

# Ordinal Encoding

```python
df4 = df1.dropna()
```

```python
df4.head()
```

```python
dict_river = {'A' : 0, 'M':1, 'O':2}
```

```python
df4['RIVER_ORDINAL'] = df4.RIVER.map(dict_river)
```
