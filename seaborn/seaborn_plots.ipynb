```python
import pandas as pd
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
iris=datasets.load_iris()
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=iris.target
```

```python
df.head()
```

```python
# Histogram 
sns.set()
sns.distplot(df['petal length (cm)'], kde=False,color='red')
```

```python
#help(sns.set())
```

```python
# KDE plot 
sns.kdeplot(data=df['petal length (cm)'], shade=True)
```

```python
sns.kdeplot([1,2,1,2,1,2,3,3,5,3,4,4,4,4,0],shade=True,color='red') #smooth histogram
```

2D KDE plots
We're not restricted to a single column when creating a KDE plot. We can create a two-dimensional (2D) KDE plot with the 
sns.jointplot command.

```python
sns.jointplot(x=df['petal length (cm)'], y=df['sepal width (cm)'], kind="kde",color='red')
```

```python
sns.set_style('dark')
# KDE plots for each species
sns.kdeplot(data=df['petal length (cm)'], label="petal", shade=True)
sns.kdeplot(data=df['sepal length (cm)'], label="sepal", shade=True)
sns.kdeplot(data=df['sepal width (cm)'], label="sepal_wid", shade=True)
```
