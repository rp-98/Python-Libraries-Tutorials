```python
n = 10
while n > 0:
    print('countdown: ', n)
    n=n-1
print('It is done. Now outside the loop.')
```

```python
n = 10
while n > 0:
    print('countdown: ', n)
    n=n-1
    if n==5:
        break
print('It is done. Now outside the loop.')
```

```python
print("This program computes n!")
n = input("Enter any positive integer: ")
num = int(n)
fact = 1
for i in range(1,num+1):
    fact = fact*i
print ('The factorial of', n, 'is', fact)
```

Recursion

```python
def f(k):
  if(k>0):
    result = k + f(k-1)
    print(result)
  else:
    result = 0
  return result

print("\n\nRecursion Results")
f(6)
```
